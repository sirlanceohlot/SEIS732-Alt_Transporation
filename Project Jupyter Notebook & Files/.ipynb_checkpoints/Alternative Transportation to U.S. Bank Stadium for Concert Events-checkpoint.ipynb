<h3>Historic_Annual_Average_Daily_Traffic_in_Minnesota.CSV (Time Series Data Source)</h3>
<br>
<b>Folder containing necessary Traffic files - Project Jupyter Notebook & Files'/Traffic_Datasets</b><br>
<br>
<b>Description</b>
<br>
Determine general estimates of location specific, annual average daily traffic (AADT) on various route systems to meet FHWA annual reporting and assist in the annual allocation of state aid to local government for roadway maintenance and construction. Counts are taken on portions of the system each year so complete coverage for a given jurisdiction is generally accomplished on a two, four, six, or twelve year cycle. Automatic traffic recorders (ATR) are permanent devices in the pavement surface that continuously and automatically collect traffic data. Currently, there are 155+ active devices maintained by MnDOT throughout the state: 75+ in the seven-county metro area and 80+ in greater Minnesota. All stations produce vehicle volume data.  However, some ATR stations collect additional types of data depending upon their equipment and sensors. In addition to collecting the type of data that ATR's capture (volume, classification, speed), weigh-in-motion (WIM) devices collect axle loadings, vehicle and axle configuration, and truck volume characteristics. <br>
<br>
<b>Rationale for Selection</b>
<br>
Historic_Annual_Average_Daily_Traffic_in_Minnesota.CSV provides the annual average daily traffic volumes of interstates and highways in Minnesota, armed with this information we are able to hone on in the Automatic Traffic Recoorders of interest leading into U.S. Bank Stadium's Downtown East neighborhood allowing the data to confirm what the Metropolitan Council already knows, traffic is constantly flowing and the Downtown East neighborhood has less stressful transportation options availablee to concert attendees.<br>

```python
# reproject the vector data to be the same as the raster
mpls_neighborhoods_df = mpls_neighborhoods_df.to_crs(crs=mpls_nlcd.crs)
```

```python
mpls_neighborhoods_df.head()
```

```python
#The neighborhood U.S. Bank Stadium resides in is Downtown East
usbank_neighborhood_df = mpls_neighborhoods_df[(mpls_neighborhoods_df['BDNAME'] == 'Downtown East')]
```

```python
usbank_neighborhood_df.head()
```

```python
usbank,out_transform = rasterio.mask.mask(mpls_nlcd, usbank_neighborhood_df['geometry'], crop=True)
```

```python
out_meta = mpls_nlcd.meta

out_meta.update({"driver": "GTiff",
                 "height": usbank.shape[1],
                 "width": usbank.shape[2],
                 "transform": out_transform})


with rasterio.open("usbank_nlcd.tif", "w", **out_meta) as dest:
    dest.write(usbank)
```

```python
# read in dataset
us_bank = rasterio.open('Geospatial_Datasets/usbank_nlcd.tif')
```

```python
band1 = us_bank.read(1)
# mask the raster data to the classifications we care about (buildings, roads)
mask = ((3 <= band1) & (band1 <= 4))

# Use the mask to extract the pixels with the target classification
masked_band = np.where(mask, band1, 0) 
```

# Filtering raw_aadt down to roads leading to U.S. Bank Stadium

```python
'''raw_aadt_df encompasses all available highway data, so we will need 
to filter down to the necessary roads leading to U.S. Bank Stadium. Utilizing the MnDOT Traffic Application
we have identified Automatic Traffic Recorders (ATR) with a Sequence Number that corresponds to the 
SEQUENCE_N column in raw_aadt_df'''

#filters to use against raw_aadt_df
years_to_filter = [2016,2017,2018,2019,2021,2022,2023]
sequences_to_filter = [9834,9838,9489,9851,9877]

filtered_aadt_df = raw_aadt_df[raw_aadt_df['DATA_YEAR'].isin(years_to_filter)]
filtered_aadt_df = filtered_aadt_df[filtered_aadt_df['SEQUENCE_N'].isin(sequences_to_filter)]
```

```python
filtered_aadt_df.head()
```

# Averaging the Annual Average of these roads leading into U.S. Bank Stadium by Year (Feature)

```python
#averaging the filtered_aadt_df by DATA_YEAR
aadt_avg_df = filtered_aadt_df.groupby(filtered_aadt_df['DATA_YEAR'])['AADT'].mean().reset_index()

#renaming the new aadt_avg_df columns
aadt_avg_df = aadt_avg_df.rename(columns={'DATA_YEAR': 'Year', 'AADT': 'Annual Avg Daily Traffic'})
```

```python
'''reviewing the averages calculated, this looks like a great feature for this dataset, we can add it to
the primary dataset, US_Bank_Concerts'''
aadt_avg_df.head()
```

# Creating an Annual Average Gas Price in real dollars (Feature)

```python
#lets adjust that column name, it is too long for this dataset
raw_gas_df = raw_gas_df.rename(columns={'Weekly Minnesota Regular Conventional Retail Gasoline Prices  (Dollars per Gallon)': 'Price Per Gallon'})

#transforming 'Date' column column to datetime
raw_gas_df['Date'] = pd.to_datetime(raw_gas_df['Date'])

#sorting by datetime
raw_gas_df.sort_values(by=['Date'], inplace=True)

#adding 'Year' column for feature design
raw_gas_df['Year'] = raw_gas_df['Date'].dt.year
#adding 'Month' colum for feature design
raw_gas_df['Month'] = raw_gas_df['Date'].dt.month

#converting 'Year' from Int32 to Int64
raw_gas_df['Year'] = raw_gas_df['Year'].astype('int32')
raw_gas_df['Month'] = raw_gas_df['Month'].astype('int32')
```

```python
# create new column linking each row to the month start date
raw_gas_df['Month Start'] = raw_gas_df.apply(lambda x: date(x['Year'], x['Month'], 1), axis=1)
```

```python
# date to convert nominal values to
dol_2023 = date(2023, 1, 1)
```

```python
#filters to use against raw_gas_df
years_to_filter = [2016,2017,2018,2019,2021,2022]
#filtering raw_gas_df
filtered_gas_df = raw_gas_df[raw_gas_df['Year'].isin(years_to_filter)]
```

```python
# convert nominal prices to real
filtered_gas_df['Price Per Gallon 2023 Dollars'] = filtered_gas_df.apply(lambda x: cpi.inflate(x['Price Per Gallon'], x['Month Start'], to=dol_2023), axis=1)
```

```python
#transforming by aggregating by year and averaging
avg_gas_df = filtered_gas_df.groupby(filtered_gas_df['Year'])['Price Per Gallon 2023 Dollars'].mean().reset_index()
```

```python
#renaming 'Price Per Gallon' column
avg_gas_df = avg_gas_df.rename(columns={'Price Per Gallon 2023 Dollars': 'Annual Avg Price Per Gallon 2023 Dollars'})
```
